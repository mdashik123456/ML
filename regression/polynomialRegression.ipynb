# Polynomial Regression

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv('../data_sets/polynomial_datasets_Ice_cream selling data.csv')
df.head()
```

```python
df.corr()
```

```python
sns.pairplot(data=df)
```

```python
y = df[['Ice Cream Sales (units)']]
x = df[["Temperature (°C)"]]
```

```python
pf = PolynomialFeatures(degree=2)
x = pf.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_train, y_train) * 100
```

```python
prd = lr.predict(x)
prd = pd.DataFrame(prd, columns=['Ice Cream Sales (units)'])
prd.head()
```

```python
plt.scatter(df["Temperature (°C)"], df["Ice Cream Sales (units)"], color='blue')
plt.plot(df["Temperature (°C)"], prd, color='red', label='Predicted')
plt.title('Ice Cream Sales Prediction')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.legend(["Orginal", "Predicted"])
```
