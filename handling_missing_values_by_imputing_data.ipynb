# handling_missing_values_by_imputing_data

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_sets/loan_data_set.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.ffill()
```

### we can use bffil() for backward fill

```python
df.bfill()
```

### for mode value

```python
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
```

“mean”, then replace missing values using the mean along each column. Can only be used with numeric data.

“median”, then replace missing values using the median along each column. Can only be used with numeric data.

 “mode”, then replace missing using the most frequent value along each column. Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.

### for mean. mean can only be use for numerical data


```python
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].mean())
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for i in df.select_dtypes(include="object").columns: # use mode for all objects columns
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
df.isnull().sum()
```

### removing all NaN values at once 

```python
numaric_cols = df.select_dtypes(exclude="object").columns # getting all numaric columns
object_cols = df.select_dtypes(include="object").columns # getting all non-numaric columns
```

```python
for object_col in object_cols:
    df[object_col] = df[object_col].fillna(df[object_col].mode()[0])
```

```python
for numaric_col in numaric_cols:
    df[numaric_col] = df[numaric_col].fillna(df[numaric_col].mode()[0])
```
