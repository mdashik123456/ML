# dropna 

dropna removes entire row when it founds any NaN value in the row. If we want to remove whole column where found NaN, we have to use axis=1 as parameter.

insted of writing `df = df.dropna()` we can write `df.dropna(inplace=True)`

```python
df.dropna(axis=1)
```

```python
df.dropna(how="any") # this will remove all NaN value row
```

```python
df.dropna(how="all") # this will remove only the row where every cell have NaN value
```

```python
df.dropna(subset=["LoanAmount"]) #Removes all the rows which have NaN value in a specific column
```

# fillna

fillna() will replace all the NaN value with given string/value

```python
df.fillna({"LoanAmount":200 , "Married" : "No", "Gender":"Male"}) # Fill the given column NaN value with give value
```

```python
df.bfill(axis=0) # ffill() -> use previous value | bfill() -> use next value | axix=0/1 premeter can be use for row/col exchange
```

# merge

```python
var1 = pd.DataFrame({"A" : [1,2,3,4,5], "B" : [11,12,13,14,15]})
var2 = pd.DataFrame({"B" : [11,12,13,14,16], "C" : [21,22,23,24,25]})

pd.merge(var1, var2, how="right") #inner outer left right
```

# groupby

```python
var = pd.DataFrame({"name" : ['a', 'b', 'a', 'c', 'b', 'a', 'd', 'd', 'c', 'a', 'b'], "s1" : [11,12,13,14,15,4,32,52,24,12,12], "s2" : [9, 10, 13, 14, 14,65,43,23,23,12, 34]})
var = var.groupby("name")
```

```python
for x,y in var:
    print(x)
    print(y)
    print("---------------")
```

```python
var.get_group("b")
```

```python
var.min() # finding minimum data accourding to name. for maximum use max()
```

# melt

melt is usedf for reshaping DataFrame

```python
var = pd.DataFrame({"days" : [1,2,3,4,5,6], "eng":[10,12,14,16,12,11], "math" : [17,15,23,12,42,23]})
var
```

```python
var.melt(id_vars=["days"])
```

# pivot

```python
var = pd.DataFrame({"days" : [1,2,3,4,5,6], "st_name":['a','b','c','a','b','c'], "eng":[10,12,14,16,12,11], "math" : [17,15,23,12,42,23]})
var
```

```python
var.pivot(index="days", columns="st_name")
```

```python
var.pivot(index="days", columns="st_name", values="eng")
```

```python
var.pivot_table(index="st_name", columns="days", aggfunc="sum", margins=True) #calculating mean/sum etc...
```
